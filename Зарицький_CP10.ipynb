```python
import requests
import zipfile
import pandas as pd
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 

%matplotlib inline

url = "http://www2.informatik.uni-freiburg.de/~cziegler/BX/BX-CSV-Dump.zip"
filename = "BX-CSV-Dump.zip"

# Завантажуємо архів
response = requests.get(url)

# Зберігаємо архів на диск
with open(filename, 'wb') as f:
    f.write(response.content)

# Розпаковуємо архів
with zipfile.ZipFile(filename, 'r') as zip_ref:
    zip_ref.extractall()

# Завантажуємо дані з файлів CSV в DataFrame
df_books = pd.read_csv('BX-Books.csv', error_bad_lines=False, delimiter=';', encoding='latin-1')
df_ratings = pd.read_csv('BX-Book-Ratings.csv', error_bad_lines=False, delimiter=';', encoding='latin-1')

# Виводимо частину датасетів на екран
print("Частина датасету BX-Books.csv:")
print(df_books.head())

print("\nЧастина датасету BX-Book-Ratings.csv:")
print(df_ratings.head())
```

### Завдання №2

* Залиште в датасеті лише колонки `ISBN`, `Book Title`,`Book Author` and `Publisher`.
* Перевірте, чи немає дублікатів назв книг, видаліть, якщо такі є. Для цього скористайтесь функціями `duplicated(subset='Book-Title')`  та `drop_duplicates(subset='Book-Title')`

* Виведіть шапку таблиці, щоб переконатись у об'єднанні



```python
# Вибираємо потрібні колонки
selected_columns = ['ISBN', 'Book-Title', 'Book-Author', 'Publisher']
df_books_selected = df_books[selected_columns]

# Видаляємо дублікати назв книг
df_books_selected = df_books_selected.drop_duplicates(subset='Book-Title')

# Об'єднуємо імена та прізвища авторів, видаляючи пробіли
df_books_selected['Book-Author'] = df_books_selected['Book-Author'].str.replace(' ', '')

# Виводимо шапку таблиці
print(df_books_selected.head())
```

### Завдання №3

* Створіть користувача, якому поставте у відповідність 5 довільних кних з переліку та рейтингом (0-10)
* Створіть рекомендований список з 10 книг для користувача, спираючись на автора та видавництво.


```python
# Створюємо словник з уподобаннями користувача
user_preferences = {
    'ISBN': ['0155061224', '0446520802', '052165615X', '0521795028', '0515107662'],
    'Rating': [8, 7, 6, 9, 5]
}

# Створюємо DataFrame з уподобаннями користувача
df_user_preferences = pd.DataFrame(user_preferences)
```

```python
# Об'єднуємо дані користувача з даними книжок за ISBN
df_user_ratings = pd.merge(df_user_preferences, df_ratings, on='ISBN')

df_user_ratings['Weighted-Rating'] = df_user_ratings['Rating'] * 0.8 + df_user_ratings['Book-Rating'] * 0.2 

# Групуємо дані за книжкою і обчислюємо середній рейтинг користувача
user_avg_ratings = df_user_ratings.groupby('ISBN')['Weighted-Rating'].mean().reset_index()

# Об'єднуємо дані користувача з середніми рейтингами книжок
df_user_books = pd.merge(user_avg_ratings, df_books_selected, on='ISBN')
```

```python
user_avg_ratings
```

```python
df_user_books = df_user_books.sort_values('Weighted-Rating', ascending=False)
df_user_books
```

```python
unique_authors = df_user_books['Book-Author'].unique()
unique_publishers = df_user_books['Publisher'].unique()

recommended_books = df_books_selected[(df_books_selected['Book-Author'].isin(unique_authors)) & 
                                     (df_books_selected['Publisher'].isin(unique_publishers))]
# Виводимо рекомендований список книжок
print("Рекомендований список книжок:")
recommended_books.head(10)
```
